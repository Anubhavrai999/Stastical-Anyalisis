```python
import matplotlib.pyplot as plt
categories=["A","B","C","D","E"]
values=[5,7,3,8,4]
width=[0.1,0.6,0.8,0.9,1]
colors=["lightgreen","blue","lightblue","red","grey"]
plt.bar(categories,values,width=width,color=colors,edgecolor="pink")
plt.xlabel("student")
plt.ylabel("age")
plt.title("madamada")
plt.grid(axis="y")
plt.show()
```

Line plot

```python
x=[12,16,7,21,8]
y=[34,23,9,11,5]
plt.plot(x,y,color="lightblue",linestyle="--",marker=">",label="datapoints")
plt.xlabel("student")
plt.ylabel("age")
plt.title("madamada")
plt.legend()
plt.grid(axis="y")
plt.show()
```

```python
x=[12,16,7,21,8]
y=[24,32,14,42,16]
plt.scatter(x,y,color="red",marker="v",label="y=x*2")
plt.xlabel("student")
plt.ylabel("age")
plt.title("Scatter plot")
plt.legend()
plt.grid(axis="y")
plt.show()
```

```python
x=[12,16,7,21,8]
y=[24,32,14,42,16]
sizes=[32,45,10,35,46]
plt.scatter(x,y,color="red",s=sizes,marker="v",label="y=x*2")
plt.xlabel("student")
plt.ylabel("age")
plt.title("Scatter plot")
plt.legend()
plt.grid(axis="y")
plt.show()
```

```python
sizes=[40,20,50,10]
labels=["grp 1","grp 2","grp 3","grp 4"]
colors=["lightgreen","blue","lightblue","red"]
plt.pie(sizes,labels=labels,autopct="%1.1f%%",startangle=90,colors=colors)
plt.title("basic pie")
plt.show()
```

```python
sizes=[40,20,50,10]
labels=["grp 1","grp 2","grp 3","grp 4"]
colors=["lightgreen","blue","lightblue","red"]
plt.pie(sizes,labels=labels,autopct="%1.1f%%",startangle=90,colors=colors,explode=(0.1,0,0,0))
plt.title("basic pie")
plt.show()
```

```python
w=0.2
x=["c1","c2","c3","c4"]
boys=[100,40,60,50]
girls=[200,50,60,90]
bar1=np.arange(len(x))
plt.bar(bar1,boys,w,label="boys")
bar2=[i+w for i in bar1]
plt.bar(bar2,girls,w,label="girls")
plt.xticks(bar1+w/2,x)
plt.show()
```

```python
w=0.2
x=["c1","c2","c3","c4"]
boys=[100,40,60,50]
girls=[200,50,60,90]
both=[10,60,30,80]
bar1=np.arange(len(x))
plt.bar(bar1,boys,w,label="boys")
bar2=[i+w for i in bar1]
plt.bar(bar2,girls,w,label="girls")
bar3=[i+2*w for i in bar1]
plt.bar(bar3,both,w,label="both")
plt.xticks(bar1+w/2,x)
plt.show()
```
